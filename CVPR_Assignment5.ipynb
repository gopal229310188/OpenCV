```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("SampleImage.jpeg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.axis("off")
plt.title("Original Color Image")
plt.show()
```

```python
def log_transformation(image):
    """
    Applies log transformation to enhance dark regions in a color image.
    """
    c = 255 / np.log(1 + np.max(image))  # Compute scaling factor
    log_image = c * np.log(1 + image.astype(np.float32))  # Apply log transformation
    log_image = np.array(log_image, dtype=np.uint8)  # Convert back to uint8

    return log_image

log_img = log_transformation(image)

plt.imshow(log_img)
plt.axis("off")
plt.title("Log Transformation (Color)")
plt.show()
```

```python
def gamma_transformation(image, gamma=1.5):
    """
    Applies power-law (gamma) transformation on a color image.
    """
    c = 255 / np.power(255, gamma)  # Compute scaling factor
    gamma_image = c * np.power(image.astype(np.float32), gamma)  # Apply gamma transformation
    gamma_image = np.array(gamma_image, dtype=np.uint8)  # Convert back to uint8

    return gamma_image

gamma_img = gamma_transformation(image, gamma=2.0)

plt.imshow(gamma_img)
plt.axis("off")
plt.title("Gamma (Power Log) Transformation (Color)")
plt.show()
```

```python
def negative_transformation(image):
    """
    Applies negative transformation to invert colors in a color image.
    """
    neg_image = 255 - image  # Subtract pixel values from 255
    return neg_image

negative_img = negative_transformation(image)

plt.imshow(negative_img)
plt.axis("off")
plt.title("Negative Transformation (Color)")
plt.show()
```

```python
image2 = cv2.imread("Nature1.jpg")
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
image2 = cv2.resize(image2, (image.shape[1], image.shape[0]))

plt.imshow(image2)
plt.axis("off")
plt.title("Second Image for Set Operations")
plt.show()
```

```python
union_img = cv2.max(image, image2)

plt.imshow(union_img)
plt.axis("off")
plt.title("Union Operation (Color)")
plt.show()
```

```python
intersection_img = cv2.min(image, image2)

plt.imshow(intersection_img)
plt.axis("off")
plt.title("Intersection Operation (Color)")
plt.show()
```
