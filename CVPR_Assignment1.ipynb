```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("/content/SampleImage.jpeg")

cv2_imshow(image)
```

```python
cv2.imwrite("/content/output_image.png", image)

converted_image = cv2.imread("/content/output_image.png")
cv2_imshow(converted_image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imwrite("/content/gray_image.jpg", gray_image)
cv2_imshow(gray_image)
```

```python
resized_image = cv2.resize(image, (100, 100))

cv2_imshow(resized_image)
```

```python
scaled_image = cv2.resize(image, None, fx=0.5, fy=0.5)

cv2_imshow(scaled_image)
```

```python
(h, w) = image.shape[:2]
center = (w // 2, h // 2)

rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated_image = cv2.warpAffine(image, rotation_matrix, (w, h))

cv2_imshow(rotated_image)
```

```python
shift_matrix = np.float32([[1, 0, 50], [0, 1, 50]])

shifted_image = cv2.warpAffine(image, shift_matrix, (w, h))

cv2_imshow(shifted_image)
```

```python
def generate_pattern(freq):
    rows, cols = 256, 256
    img = np.zeros((rows, cols), dtype=np.uint8)
    for i in range(rows):
        for j in range(cols):
            img[i, j] = 255 * (np.sin(2 * np.pi * freq * (i + j)) > 0)
    return img

freq = float(input("Enter frequency: "))

pattern_img = generate_pattern(freq)
cv2_imshow(pattern_img)
```
