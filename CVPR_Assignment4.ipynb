```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("SampleImage.jpeg", cv2.IMREAD_GRAYSCALE)

plt.imshow(image, cmap="gray")
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
def log_transformation(image):
    """
    Applies log transformation to enhance dark regions.
    """
    c = 255 / np.log(1 + np.max(image))
    log_image = c * np.log(1 + image.astype(np.float32))
    log_image = np.array(log_image, dtype=np.uint8)

    return log_image

log_img = log_transformation(image)

plt.imshow(log_img, cmap="gray")
plt.axis("off")
plt.title("Log Transformation")
plt.show()
```

```python
def gamma_transformation(image, gamma=1.5):
    """
    Applies power-law (gamma) transformation.
    """
    c = 255 / np.power(255, gamma)
    gamma_image = c * np.power(image.astype(np.float32), gamma)
    gamma_image = np.array(gamma_image, dtype=np.uint8)

    return gamma_image

gamma_img = gamma_transformation(image, gamma=2.0)

plt.imshow(gamma_img, cmap="gray")
plt.axis("off")
plt.title("Gamma (Power Log) Transformation")
plt.show()
```

```python
def negative_transformation(image):
    """
    Applies negative transformation to invert colors.
    """
    neg_image = 255 - image
    return neg_image

negative_img = negative_transformation(image)

plt.imshow(negative_img, cmap="gray")
plt.axis("off")
plt.title("Negative Transformation")
plt.show()
```

```python
image2 = cv2.imread("Nature1.jpg", cv2.IMREAD_GRAYSCALE)
image2 = cv2.resize(image2, (image.shape[1], image.shape[0]))

plt.imshow(image2, cmap="gray")
plt.axis("off")
plt.title("Second Image for Set Operations")
plt.show()
```

```python
union_img = cv2.max(image, image2)

plt.imshow(union_img, cmap="gray")
plt.axis("off")
plt.title("Union Operation")
plt.show()
```

```python
intersection_img = cv2.min(image, image2)

plt.imshow(intersection_img, cmap="gray")
plt.axis("off")
plt.title("Intersection Operation")
plt.show()
```

```python
difference_img = cv2.absdiff(image, image2)

plt.imshow(difference_img, cmap="gray")
plt.axis("off")
plt.title("Difference Operation")
plt.show()
```
